````python
def welcome():
    nombre = input("Dime tu nombre\n")
    year = input("Dime el año en que naciste\n")
    print(f"Hola {nombre}, tienes {2020 - int(year)} años.")
````

````python
welcome()
````

````python
try:
    welcome()
except:
    print("Algo ha fallado")
````

````python
try:
    welcome()
except Exception as e:
    print("Algo ha fallado:")
    print(e)
````

````python
while True:
    try:
        welcome()
        break
    except Exception as e:
        print("Algo ha fallado:")
        print(e)
        print("Intentalo otra vez")

print("Adios")
````

````python
def division(a,b):
    try:
        return a/b
    except TypeError:
        print("Tipo invalido. a y b deben ser int o float")
    except ZeroDivisionError:
        print("No dividas por cero! Que haces, loco?")
    except Exception as e:
        print("Ha fallado, verifique error")
        print(e)
````

## Defensive Programming

````python
def division(a,b):
    if type(a) != int:
        raise ZeroDivisionError("a no es un entero")
    return a/b
````

````python
try:
    division("i",5)
except Exception as e:
    print("Algo ha fallado")
    print(type(e))
````

````python
def division(a,b):
    if type(a) != int or type(a) != float:
        try:
            a = float(a)
        except:
            raise TypeError(f"a no es compatible, debe ser un entero o float y es del tipo {type(a)}")
    return a/b
````

````python
def division(a,b):
    if type(a) != int or type(a) != float:
        try:
            a = float(a)
        except:
            raise TypeError(f"a no es compatible, debe ser un entero o float y es del tipo {type(a)}")
    if type(b) != int or type(b) != float:
        try:
            b = float(b)
        except:
            raise TypeError(f"b no es compatible, debe ser un entero o float y es del tipo {type(b)}")
    if b == 0:
        raise ZeroDivisionError("b no puede ser 0")
    if a == 75.32:
        raise Exception("Libreria de Felipe: No me gusta el numero 75.32, por favor no lo uses\n Mirad la documentacion en http://www.felipe.com")
    return a/b
````
